```python
import requests
import json
import time
import pandas as pd
import random
from sqlalchemy import create_engine
```

```python
headers = {
    "X-RapidAPI-Host": "amazon-products1.p.rapidapi.com",
    "X-RapidAPI-Key": "API_KEY"
}
```

```python
asin = []
title = []
link = []
review_count = []
rating = []
current_price = []
previous_price = []
currency = []
is_prime = []
is_sponsored = []
is_amazon_choice = []
```

```python
url_search = 'https://amazon-products1.p.rapidapi.com/search'

search = ["clothing", "electronics", "appliance", "sport", "camping", "baking", "toy"]

for query in search:
    for page in range(1,3):
        querystring = { 
            "country":"US",
            "query":query,
            "page": page
        }

        request = requests.get(url_search, headers = headers, params = querystring)
        
        result = request.json()
        
        for product in result["results"]:
            asin.append(product["asin"])
            title.append(product["title"])
            link.append(product["full_link"])
            review_count.append(product["reviews"]["total_reviews"])
            rating.append(product["reviews"]["stars"])
            current_price.append(product["prices"]["current_price"])
            previous_price.append(product["prices"]["previous_price"])
            currency.append(product["prices"]["currency"])
            is_prime.append(product["prime"])
            is_sponsored.append(product["sponsored"])
            is_amazon_choice.append(product["amazon_choice"])
        
        time.sleep(2)
```
